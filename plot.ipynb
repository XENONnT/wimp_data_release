```python
import matplotlib.pyplot as plt
import pandas as pd
```

### SR0

```python
def make_plot(df):
    mass = df['WIMP mass [GeV/c^2]']
    limit = df['unconstrained upper limit [cm^2]']
    limit_pcl = df['upper limit [cm^2]']
    sensi_med = df['median upper limit [cm^2]']
    sensi_pos_1sig = df['1sigmath sensitivity quantile [cm^2]']
    sensi_pos_2sig = df['2sigmath sensitivity quantile [cm^2]']
    sensi_neg_1sig = df['-1sigmath sensitivity quantile [cm^2]']
    sensi_neg_2sig = df['-2sigmath sensitivity quantile [cm^2]']

    ax = plt.subplot()
    ax.fill_between(mass, sensi_neg_2sig, sensi_pos_2sig, color='#FDED95')
    ax.fill_between(mass, sensi_neg_1sig, sensi_pos_1sig, color='#83C369')
    ax.plot(mass, sensi_med, color='w', lw=0.3)
    ax.plot(mass, limit, ls='--', color='k')
    ax.plot(mass, limit_pcl, color='k')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('WIMP mass [GeV/c$^2$]')
    ax.set_ylabel('WIMP-nucleon cross section [cm$^2$]')
    ax.set_xlim(mass.min(), mass.max())

    return ax
```

```python
df = pd.read_csv('SR0/XENONnT_2023_WIMP_SI.csv')
ax = make_plot(df)
ax.set_title('Spin-independent WIMP-nucleon interaction')
plt.show()

df = pd.read_csv('SR0/XENONnT_2023_WIMP_SDn.csv')
ax = make_plot(df)
ax.set_title('Spin-dependent WIMP-nucleon interaction (neutron-only)')
plt.show()

df = pd.read_csv('SR0/XENONnT_2023_WIMP_SDp.csv')
ax = make_plot(df)
ax.set_title('Spin-dependent WIMP-nucleon interaction (proton-only)')
plt.show()
```

### SR0+SR1

```python
df = pd.read_csv('SR01/xenonnt_2025_si_wimp.csv')
```

```python
def make_plot(df):
    mass = df['mass']
    limit = df['upper_limit']
    limit_pcl = df['upper_limit']
    sensi_med = df['sensitivity_0']
    sensi_pos_1sig = df['sensitivity_1']
    sensi_pos_2sig = df['sensitivity_2']
    sensi_neg_1sig = df['sensitivity_-1']
    sensi_neg_2sig = df['sensitivity_-2']

    ax = plt.subplot()
    ax.fill_between(mass, sensi_neg_2sig, sensi_pos_2sig, color='#FDED95')
    ax.fill_between(mass, sensi_neg_1sig, sensi_pos_1sig, color='#83C369')
    ax.plot(mass, sensi_med, color='w', lw=0.3)
    ax.plot(mass, limit, ls='--', color='k')
    ax.plot(mass, limit_pcl, color='k')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('WIMP mass [GeV/c$^2$]')
    ax.set_ylabel('WIMP-nucleon cross section [cm$^2$]')
    ax.set_xlim(mass.min(), mass.max())

    return ax
```

```python
df = pd.read_csv('SR01/xenonnt_2025_si_wimp.csv')
ax = make_plot(df)
ax.set_title('Spin-independent WIMP-nucleon interaction')
plt.show()

df = pd.read_csv('SR01/xenonnt_2025_sd_neutron_wimp.csv')
ax = make_plot(df)
ax.set_title('Spin-dependent WIMP-nucleon interaction (neutron-only)')
plt.show()

df = pd.read_csv('SR01/xenonnt_2025_sd_proton_wimp.csv')
ax = make_plot(df)
ax.set_title('Spin-dependent WIMP-nucleon interaction (proton-only)')
plt.show()
```
